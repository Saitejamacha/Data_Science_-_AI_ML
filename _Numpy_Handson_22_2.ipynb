```python
#zero dimension
a_0=np.array(5)
#shape()
# 0-()
# 1-(a,)- no of elements in the array
# 2-(a,b)
# 3-(a,b,c)

print(a_0)
print(a_0.shape)
print(a_0.ndim)
```

```python
#1 dimension
a_1=np.array([2,4,2,1,3],dtype='float')

print(a_1)
print(a_1.shape)
print(a_1.ndim)
```

```python
#2 dimension
a_2=np.array([[1,2,3,4,5],[10,9,10,8,7]])
print(a_2)
print(a_2.shape)
# 2-(a,b)
#a-no of 1d array
#b-no of values in each 1d array
print(a_2.ndim)
```

```python
#3 dimension
a_3=np.random.randint(10,20,(3,3,5))
print(a_3)
print(a_3.shape)
# 3-(a,b,c)
#a- no of 2d array
#b-no of 1d array in each 2d array
#c-no of values in each 1d array
print(a_3.ndim)
```

```python
10. Given a numpy array that has values = [1,2,3,4,5]. Perform the following operations.
a. Insert the element 6 at the end of the array.
b. Insert the element 10 at the beginning of the array.
c. Insert the element 30, before 4.
d. Update the element 5 as 50.
e. Delete the element 50 in the updated array.
f. Create a function to search for the element 30.
g. Sort the array in ascending order.
```

```python
arr=np.array([1,2,3,4,5])
arr
```

```python
# a. Insert the element 6 at the end of the array.
arr=np.append(arr,6)
```

```python
arr
```

```python
# b. Insert the element 10 at the beginning of the array.
arr=np.insert(arr,0,10)
```

```python
arr
```

```python
# c. Insert the element 30, before 4.
np.where(arr==4)[0][0]
```

```python
arr=np.insert(arr,np.where(arr==4)[0][0],30)
arr
```

```python
# d. Update the element 5 as 50.
np.where(arr==5)[0][0]
```

```python
arr[6]=50
```

```python
arr
```

```python
# e. Delete the element 50 in the updated array.
np.delete(arr,np.where(arr==50)[0][0])
```

```python
f. Create a function to search for the element 30.
g. Sort the array in ascending order.
```

```python
def search(arr,value):
  for i in arr:
    if i==value:
      print(f"{value}: is found")
      break
  else:
     print(f"{value} is not found")
search(arr,30)
```

```python
#sorting
np.sort(arr)
```
